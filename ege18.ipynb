Ответ: 1204

````python
data = [list(map(int, s.split(';'))) for s in open("DATA/18-0.csv")]
rows = len(data)
cols = len(data[0])

work = data

for col in range(cols-2,-1,-1):
  work[rows-1][col] = data[rows-1][col] + work[rows-1][col+1]

for row in range(rows-2,-1,-1):
  work[row][cols-1] = data[row][cols-1] + work[row+1][cols-1]

for row in range(rows-2,-1,-1):
  for col in range(cols-2,-1,-1):
     work[row][col] = data[row][col] + max(work[row][col+1], work[row+1][col])

print(work[0][0])
````
